```python
import pandas as pd
import numpy as np
```

```python
domicilios = pd.read_csv('DM_DatosDomicilio.txt',delimiter='|')
```

```python
domicilios
```

```python
colonias = pd.read_csv('colonias.csv')
```

```python
colonias
```

```python
domicilios.merge(colonias, on=['FC_ID_ESTADO','FC_CODIGOPOSTAL','FC_COLONIAHOGAR'], how='left')
```

```python
m_domicilios = domicilios.merge(colonias, on=['FC_ID_ESTADO','FC_CODIGOPOSTAL','FC_COLONIAHOGAR'], how='left')
```

```python
m_domicilios
```

```python
m_domicilios = m_domicilios[['#FI_ID_INSTITUCION','FI_EMPLEADO','FC_CALLEHOGAR','FC_NUMEXTHOGAR','FC_NUMINTHOGAR','FC_COLONIAHOGAR','FC_MUNICIPIOHOGAR','FC_ESTADOHOGAR','FC_CODIGOPOSTAL','FC_PAIS','FI_ID_TIPOVIVIENDA','FC_ID_TIPOVIVIENDA','FC_LATITUD','FC_LONGITUD','FC_SUCURSAL','FI_ID_COLONIA_y','FI_ID_MUNICIPIO_x','FC_ID_ESTADO','FI_ID_PAIS_x']]
```

```python
m_domicilios['FI_ID_COLONIA_y'] = m_domicilios['FI_ID_COLONIA_y'].fillna(-1)
```

```python
m_domicilios['FI_ID_COLONIA_y'] = m_domicilios['FI_ID_COLONIA_y'].astype(int)
```

```python
m_domicilios
```

```python
m_domicilios.to_csv('DM_DatosDomicilio.txt.2',sep='|',index=False)
```
